## 2. IRIS² Hybrid Constellation Simulation
## 2. Simulación de la Constelación Híbrida IRIS²

IRIS² relies on a hybrid architecture combining:
- **LEO Satellites** (Low Earth Orbit, 600 - 800 km) for secure, high-throughput, low-latency communication.
- **MEO Satellites** (Medium Earth Orbit, ~20,000 km) for wider coverage and navigation linkage.

```python
# Model IRIS² planes / Modelar los planos de IRIS²
# We design 3 LEO planes with different RAAN and 1 MEO plane

leo_plane_1 = create_leo_orbit(altitude=650, inclination=85.0, raan=0.0)
leo_plane_2 = create_leo_orbit(altitude=650, inclination=85.0, raan=120.0)
leo_plane_3 = create_leo_orbit(altitude=650, inclination=85.0, raan=240.0)
meo_plane   = create_meo_orbit(altitude=21500, inclination=56.0, raan=0.0)

plotter = OrbitPlotter()
fig = plotter.plot_comparison(
    [leo_plane_1, leo_plane_2, leo_plane_3, meo_plane],
    names=['LEO Plane A', 'LEO Plane B', 'LEO Plane C', 'MEO Plane (Galileo sync)'],
    colors=['#00ff88', '#00bfff', '#ffd700', '#ff1493'],
    title='IRIS² Hybrid Constellation Architecture'
)

# Record constellation design achievement
unlocked = tracker.record_action("model_iris2")
for ach in unlocked:
    print(f"UNLOCKED LOGRO: {ach.id} - {ach.tier.upper()}!")

fig.show()
```
